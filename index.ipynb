## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
- One-to-many: Means the table with the primary key holds the foreign key for the other table. This makes it possible for the primary table to contain multiple foreign keys for other tables. e.g. one customer having multiple orders
- Many-to-many: Relationship where a table can hold many keys for other tables, and those tables can in turn have several keys linking back to the primary table. 

## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
1. Inner join: The most common. Joins together two tables, using the specified index, and the new table contains all the rows that have matching index. 
2. Left Join: The new table will have all the data in left table as well as any data that has matching keys in the right table. 
3. Right Join: Opposite of Left join.
4. Outer Join: Creates new table with all rows from both tables.
